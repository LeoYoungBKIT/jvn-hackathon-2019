```python
import pandas as pd
```

```python
df = pd.read_excel("Vietnam-Macroeconomic-Data.xls", header=None)
```

```python
df = df.transpose()
```

```python
df = df.rename(columns={0:'Year', 1:'GDP', 2:'UR'}).drop(df.index[0])
```

```python
df['UR'] = df['UR'].map(lambda x: -1 if x == 'no data' else float(x))
```

```python
df['Year'] = df['Year'].astype(int)
df['GDP'] = df['GDP'].astype(float)
```

```python
df = df.set_index('Year')
```

```python
nonna_df = df[df['UR'] != -1]
```

```python
nonna_df.cov()
```

```python
nonna_df.corr()
```

## Predict missing value

```python
from sklearn.linear_model import LinearRegression
```

```python
X = nonna_df['GDP'].values.reshape(-1,1)
y = nonna_df['UR'].values.reshape(-1,1)
```

```python
linreg = LinearRegression()
linreg.fit(X, y)
```

```python
missing_ur = df[df['UR'] == -1]['GDP'].values.reshape(-1,1)
```

```python
predict_ur = linreg.predict(missing_ur).reshape(-1)
```

```python
predict_ur
```

```python
# fill missing value

for i in range(len(predict_ur)):
    df.iloc[i]['UR'] = predict_ur[i]
```

## Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# GDP

idx = range(len(df))
plt.figure(figsize=(25,10))
plt.bar(idx, df['GDP'])
plt.xticks(idx, df.index)
plt.xlabel("Year")
plt.ylabel("GDP")
plt.title("GDP, current prices (Billions of U.S. dollars)")
plt.show()
```

```python
# Unemployment rate

idx = range(len(df))
plt.figure(figsize=(25,10))
plt.bar(idx, df['UR'])
plt.xticks(idx, df.index)
plt.xlabel("Year")
plt.ylabel("Unemployment rate")
plt.title("Unemployment rate (Percent)")
plt.show()
```
